```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut 
```

```python
file_to_load = "steam_sales_by_country.csv"
lat_lng = "world_country_and_usa_states_latitude_and_longitude_values.csv"
```

```python
steam_data = pd.read_csv(file_to_load)
steam_data.head()
```

```python
lat_lng_data = pd.read_csv(lat_lng)
lat_lng_data.head()
```

```python
merge_df = pd.merge(lat_lng_data, steam_data, on=["Country"])
```

```python
merge_df.head()
```

```python
merge_df.to_csv('merge.csv')
```
